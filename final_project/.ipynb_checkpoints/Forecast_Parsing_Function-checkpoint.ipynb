Предсказание средней температуры

```python
import requests      # Библиотека для отправки запросов
import numpy as np   # Библиотека для матриц, векторов и линала
import pandas as pd  # Библиотека для табличек 
import time          # Библиотека для времени
```

```python
from fake_useragent import UserAgent
```

```python
from bs4 import BeautifulSoup
```

```python
def getAvarageTemperature(city_code, year, mounth):
    try:
        url = 'http://www.pogodaiklimat.ru/monitor.php?id={}&month={}&year={}'.format(city_code, mounth, year)
        response = requests.get(url, headers={'User-Agent': UserAgent().chrome})
        html = response.content
        soup = BeautifulSoup(html, 'html.parser') 
        obj = soup.find_all(lambda tag: tag.name == 'div' and tag.get('class') == ['climate-text'])[1]
        temperature = obj.find_all(lambda tag: tag.name == 'b')[1].contents[0]
        temperature = float(temperature[:-1])
    except:
        temperature=None
    return temperature    
```

```python
def getYear(city_code, year):
    year_temp = []
    for i in range(12):
        temperature = getAvarageTemperature(city_code, year, i+1)
        year_temp.append(temperature) 
        time.sleep(0.3)
    return year_temp
```

```python
def getCity(city_code, start_year, finish_year):
        years = finish_year - start_year + 1
        city_temp = []
        for i in range(years):
            temperature = getYear(city_code, start_year+i)
            city_temp = city_temp + temperature
        return city_temp
```

```python
start_year = 2010
finish_year = 2012

city_codes = [27402, 26063, 27532, 28748, 27719, 27612]
```

```python
final_temp = []

for city_code in city_codes:
    print(city_code)
    city_temp = getCity(city_code, start_year, finish_year)
    final_temp.append(city_temp) 

final_temp
```

```python
final_df = pd.DataFrame(final_temp).T
final_df.columns=["Тверь", "Санкт-Петербург", "Владимир", "Троицк", "Тула", "Москва"]
```
